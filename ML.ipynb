```python
import pandas as pd
df=pd.read_csv("/content/Generative AI Tools - Platforms 2025.csv")
df.head()
```

```python
X=df.iloc[:,:]
print(X)
y=df.iloc[:,-1]
y
```

```python
x_train.shape
```

```python
from sklearn.preprocessing import MinMaxScaler

mm=MinMaxScaler()
# Select only numeric columns before fitting the scaler
numeric_cols = x_train.select_dtypes(include=['number']).columns
mm.fit(x_train[numeric_cols])
x_train=mm.transform(x_train[numeric_cols])
```

```python
x_train
```

```python
# Select only numeric columns before transforming the test data
numeric_cols = x_test.select_dtypes(include=['number']).columns
x_test=mm.transform(x_test[numeric_cols])
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
# Use the scaled numeric data for training
knn.fit(x_train,y_train)
```

```python
y_predict=knn.predict(x_test)
```
